# Data Analysis

```python
def print_highest_statistic(df, column_name, column_label):
    player_name = df.loc[df[column_name].idxmax(), 'player']
    team_name = df.loc[df[column_name].idxmax(), 'team']
    statistic_value = df.loc[df[column_name].idxmax(), column_name]
    opponent = df.loc[df[column_name].idxmax(), 'opponent']

    print(f"----> Highest {column_label} were achieved by the player '{player_name}' "
          f"from the team '{team_name}' with a value of {statistic_value} against team {opponent}")

print_highest_statistic(df, 'runs', 'runs')
print_highest_statistic(df, 'wickets', 'wickets')
print_highest_statistic(df, '6s', '6s')
print_highest_statistic(df, '4s', '4s')
print_highest_statistic(df, 'maidens', 'maidens')
```

```python
# group the players by the overs, runs, wickets.....

df_player_overs=df.groupby('player').agg({'overs':'sum', 'maidens':'sum', 'runs':'sum', 'wickets':'sum', 'run_rate':'mean', '4s':'sum', '6s':'sum', 'wd':'sum'}).reset_index()
```

```python
# sort the dataset by the overs
sorted_df_player_overs = df_player_overs.sort_values(by='overs', ascending=False).reset_index()
```

```python
sorted_df_player_overs
```

```python
# overs by the top 15 players

df_player_overs_top_15=sorted_df_player_overs.head(15)

# Plotting
plt.figure(figsize=(15, 7))
plt.bar(df_player_overs_top_15['player'], df_player_overs_top_15['overs'])
plt.xlabel('players')
plt.ylabel('overs')
plt.title('overs by the top 15 players')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()


# Show the plot
plt.show()

```

```python
# Highest runs given by top 10 bowlers

df_player_overs_top_10=sorted_df_player_overs.head(10)

# Plotting
plt.figure(figsize=(15, 7))
plt.bar(df_player_overs_top_10['player'], df_player_overs_top_10['overs'])
plt.xlabel('players')
plt.ylabel('overs')
plt.title('overs by the top 10 players')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()


# Show the plot
plt.show()
```

```python
# Maiden overs and Overs of top 10 players

fig, ax1 = plt.subplots(figsize=(15, 7))

# Plot bars for 'overs' on the primary y-axis (ax1)

index = np.arange(len(df_player_overs_top_10))
bars = ax1.bar(index, df_player_overs_top_10['overs'], label='Overs', color='skyblue')

# Plot line for 'wickets' on the same y-axis (ax1)
line = ax1.plot(index, df_player_overs_top_10['maidens'],  marker='o', label='maidens')

# Set labels and titles
ax1.set_xlabel('Player')
ax1.set_ylabel('Overs')

plt.title('Maiden overs and Overs of top 10 players')

# Combine legends
bars_legend = [bar for bar in bars]
line_legend = [line[0]]
plt.legend(bars_legend + line_legend, ['Overs', 'Maiden '], loc='upper left')

# Set x-axis ticks and labels
plt.xticks(index, df_player_overs_top_10['player'], rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Run_rate by the Team bowlers

plt.figure(figsize=(15, 7))

# Plot the data points
plt.plot(df_team_overs['team'], df_team_overs['run_rate'], marker='o', linestyle='-')

# Annotate each data point with its corresponding value (formatted to 2 decimal places)
for team, run_rate in zip(df_team_overs['team'], df_team_overs['run_rate']):
    plt.text(team, run_rate, f'{run_rate:.2f}', ha='center', va='bottom')

plt.xlabel('Teams')
plt.ylabel('Run Rate')
plt.title('Run Rate by the Team bowlers')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()

# Show the plot
plt.show()

```

```python
# Wickets taken by Each Team

plt.figure(figsize=(5, 5))
plt.pie(df_team_overs['wickets'], labels=df_team_overs['team'], autopct='%1.1f%%', startangle=90)
plt.title('Wickets taken by Each Team')
# plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

plt.show()
```

```python
# Runs given by the Team bowlers

plt.figure(figsize=(15, 7))

# Plot the data points
plt.plot(df_team_overs['team'], df_team_overs['runs'], marker='o', linestyle='--')

# Annotate each data point with its corresponding value
for team, runs in zip(df_team_overs['team'], df_team_overs['runs']):
    plt.text(team, runs, str(runs), ha='center', va='bottom',)

plt.grid()
plt.gray()
plt.xlabel('Teams')
plt.ylabel('Runs')
plt.title('Runs given by the Team bowlers')
plt.xticks(rotation=45, ha='right')  # Rotate player names for better visibility
plt.tight_layout()

# Show the plot
plt.show()
```
